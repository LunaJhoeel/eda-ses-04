Abrimos el archivo en un dataframe como .parquet porque conserva el tipo de dato a diferencia de .csv que convierte todo a string

Revisamos una muestra del DataFrame para evaluar si remover el decimal o no

Como todos parecen ser ceros y decimales, estos se eliminarán

```python
def eliminar_punto_cero(valor):
    try:
        numero = float(valor)
        if numero == int(numero):
            return str(int(numero))
        else:
            return str(numero)
    except:
        return valor

df_suministro['NRO_SERVICIO'] = df_suministro["NRO_SERVICIO"].apply(eliminar_punto_cero)
```

Convertimos la columna de clientes

```python
df_suministro["NRO_SERVICIO"] = df_suministro["NRO_SERVICIO"].astype(str)
df_suministro.head(3)
```

A partir de los nombres de las columnas, se puede inferir que se tratan de periodos de tiempo. Revisamos la cantidad de valores nulos por cliente

```python
df_suministro.isna().sum(axis=1)
```

Se detecta la presencia de valores faltantes por cada cliente vamos a ordenarlos de mayor a menor

```python
df_suministro.isna().sum(axis=1).sort_values(ascending=False)
```

Se puede inferir de que algunos suministros tienen valores incompletos porque existe la posibilidad de que aún no hayan sido instalados

```python
# Selecciona columnas de consumo
columnas_consumo = []
for columna in df_suministro.columns:
    if columna.startswith('consumo_'):
        columnas_consumo.append(columna)
columnas_consumo = sorted(columnas_consumo)

df_consumo = df_suministro[columnas_consumo].copy()

df_muestra = df_consumo.sample(n=10, random_state=42)

# Plot
plt.figure(figsize=(16, 8))
sns.heatmap(df_muestra, cmap="YlGnBu", linewidths=0.5, linecolor='gray', mask=df_muestra.isna())

plt.title('Mapa de consumo: muestra aleatoria de 10 suministros')
plt.xlabel('Mes (AAAAMM)')
plt.ylabel('Suministro')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Se confirma de que algunos suministros aun no registraban consumo en periodos previos por lo que se puede suponer que los medidores aun no se encontraban instalados. 

```python
consumo_total_crudo = df_consumo.sum(axis=0)

plt.figure(figsize=(6, 6))
sns.boxplot(y=consumo_total_crudo.to_numpy(), color='tomato')
plt.title('Consumo total (sin imputar)')
plt.ylabel('Consumo total (kWh)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Sin imputar los valores, visualmente no se detecta la presencia de valores atipicos (outliers) puesto que el consumo total del sistema se ve estable. Sin embargo, se  procedera a imputar los valores en las series de tiempo para analizar series completas

### Imputacion de valores

La imputacion de valores faltantes para este caso corresponde a rellenarlo con la media desde el periodo (T0) de consumo de la siguiente manera

```python
df_imputado = df_consumo.copy()

for i in df_imputado.index:
    fila = df_imputado.loc[i]

    # Encuentra el primer periodo con consumo
    primer_valido = None
    for columna in columnas_consumo:
        if pd.notna(fila[columna]):
            primer_valido = columna
            break

    # Si se encontro, imputa interpolando desde ahi
    if primer_valido is not None:
        datos_a_imputar = fila.loc[primer_valido:]
        datos_imputados = datos_a_imputar.interpolate(method='linear', limit_direction='both')
        df_imputado.loc[i, primer_valido:] = datos_imputados
```

A continuacion, se agregan los valores para tener un consumo total

```python
df_agregado = df_imputado[columnas_consumo].copy()

# Suma valores por mes
consumo_total_mensual = df_agregado.sum(axis=0)

# REmueve la parte 'consumo_' del nombre de la columna
etiquetas_meses = []
for col in columnas_consumo:
    etiquetas_meses.append(col.replace("consumo_", ""))
```

### Deteccion de Outliers

```python
df_box = pd.DataFrame({
    'Mes': etiquetas_meses,
    'ConsumoTotal': consumo_total_mensual.to_numpy()
})

# Plot tipo caja
plt.figure(figsize=(6, 6))
sns.boxplot(data=df_box, y='ConsumoTotal', color='lightblue')

plt.title('Consumo total mensual para detección de outliers')
plt.ylabel('Consumo total (kWh)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

En la visualizacion se puede observar que el consumo tipico esta por los 4.7 millones de kWh en el periodo completo, la caka indica que el consumo se encuentra entre 4.3 y 5.2 millones de kWh y parece que los con la imputacion por interpolacion tampoco existen valores atipicos
